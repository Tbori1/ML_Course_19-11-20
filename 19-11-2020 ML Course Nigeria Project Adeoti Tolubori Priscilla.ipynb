```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBRFClassifier
from sklearn.neighbors import KNeighborsClassifier
from xgboost import XGBClassifier
from sklearn.metrics import recall_score, accuracy_score, precision_score, f1_score, confusion_matrix
```

```python
model = LogisticRegression(solver = 'liblinear',random_state = 42)
```

```python
model.fit(X_train, y_train)
```

```python
print("Accuracy of train:",model.score(X_train,y_train))
print("Accuracy of test:",model.score(X_test,y_test))
```

```python
y_pred = model.predict(X_test)
confusion_matrix(y_test, y_pred)
```

```python
print("Precision = {}".format(precision_score(y_test, y_pred)))
print("Recall = {}".format(recall_score(y_test, y_pred)))
print("Accuracy1 = {}".format(accuracy_score(y_test, y_pred)))
print("F1 Score = {}".format(f1_score(y_test, y_pred)))
```

```python
#DecisionTreeClassifier
clf = DecisionTreeClassifier(max_depth=4, random_state=42)
clf.fit(X_train,y_train)
print("Accuracy of train:",clf.score(X_train,y_train))
print("Accuracy of test:",clf.score(X_test,y_test))
```

```python
from sklearn.metrics import precision_score, recall_score, accuracy_score, classification_report, f1_score
pred = clf.predict(X_test)
print(classification_report(y_test,pred))
```

```python
print("Precision = {}".format(precision_score(y_test, pred)))
print("Recall = {}".format(recall_score(y_test, pred)))
print("Accuracy = {}".format(accuracy_score(y_test, pred)))
print("F1 Score = {}".format(f1_score(y_test, pred)))
```
